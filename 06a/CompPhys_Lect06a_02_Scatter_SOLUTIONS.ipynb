# SOLUTIONS - Advanced Plotting: Scatter plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

Basic scatter plots are fairly straightforward.

Say we want to show that we have measured the gravitational acceleration at sea level for different positions on the a planet

```python
l = np.array([3.98850818, 0.49816757, 5.6971237,  1.51783996, 4.05110474, 3.29440836,
 1.70522641, 3.82721614, 0.48209645, 5.56109713])
g = np.array([3.79690656, 3.78591154, 3.91298875, 3.79202157, 3.77546206, 3.80090224,
 3.81505039, 3.72326755, 3.67813356, 3.79541003])

plt.scatter(l, g)
plt.xlabel("Longitude, $\lambda$ [radians]")
plt.ylabel("Gravitational acceleration, $g$ [ms$^{-1}$]")
plt.grid()
```

We can add different scatters to the same plot for different data series, say you take measurements using different methods and want to display that they are consistent.

```python
l2 = np.array([1.78304202, 4.18965029, 1.81093636, 2.39791122, 2.54620499, 6.27186422, 1.77541179, 0.32513906, 5.85135491, 2.44024172])
g2 = np.array([3.65079601, 3.78076687, 3.61996603, 3.83169851, 3.85492977, 3.80159036, 3.67007362, 3.66771254, 3.74172093, 3.39304574])

plt.scatter(l, g, label="Method 1")
plt.scatter(l2, g2, label="Method 2")
```

Things can get difficult when, for example, you have lots of scatter points, and you can't see the structure

```python
l_lots = np.random.uniform(size=20000) * 2 * np.pi
g_lots = np.random.normal(loc=3.76, scale=0.1, size=20000)

plt.scatter(l_lots, g_lots)
```

We can combat this in a few ways
* color of the datapoints
* how see-through the points are
* size of the data points

```python
# Add a colour parameter of how far away from the mean parameter the data points are
z = (abs(g_lots - g_lots.mean()) / g_lots.std())

plt.scatter(l_lots, g_lots, c=z, cmap='viridis_r')

# Don't forget to include details of what the color means!
cb = plt.colorbar(label="Distance from mean, standard deviations")
```

We may not have any information to plot on the z-axis though, so we can use the transparency of the datapoints to provide more information.
This acts as a proxy for the density of datapoints, as the more overlapping points there are, the darker the plot is. We also make the data points smaller for this.

```python
plt.scatter(l_lots, g_lots, alpha=0.1, s=5)
```

## Errors

Usually, we will have known errors on our measurements, and so we can include that information.

For this, we use the `errorbar` plot; this is

```python
# Some random errors on our datapoints
e = np.array([0.04632176, 0.04313373, 0.04390741, 0.04097481, 0.04216593,
       0.04320637, 0.04559787, 0.04212081, 0.04648771, 0.04708792])

plt.errorbar(l, g, yerr=e, fmt='o')
plt.xlabel("Longitude, $\lambda$ [radians]")
plt.ylabel("Gravitational acceleration, $g$ [ms$^{-1}$]")
plt.grid()
```
