## Distributions and Fitting a model to the data

1. Create a Gaussian distrbituion with a mean of 10 and a standard deviation of 3

2. Add some noise which follows a uniform distribution to this data and plot it. Hint, use the max value of the Gaussian distribution as an upper boundary in your uniform distribution

3. Using scipy.curve_fit, find a model which fits the data in question 2 well.

```python
# Answer 1
import scipy.stats as stats

x = np.arange(0, 20, 0.01)
y = stats.norm.pdf(x, 10, 3)
plt.plot(x,y)
```

```python
# Answer 2
noise_d = np.random.uniform(0,max(y),len(y))

y_noise = y + noise_d
plt.plot(x,y_noise)
```

```python
# Answer 3
from scipy.optimize import curve_fit

# Define Gaussian and use mean and std given in question
def gauss(x, a, b):
  return a + b * np.exp(-(x - 10) ** 2 / (2 * 3 ** 2))

params, _ = curve_fit(gauss, x, y_noise)

print(params)

model_y = gauss(x, params[0], params[1])

plt.plot(x,y_noise)
plt.plot(x,model_y, c='red')
```
