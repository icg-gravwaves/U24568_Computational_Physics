# Calculating g with errors

Imagine you are in a lab. You are measuring g by rolling a ball down a 1m slope.

You will use:

$$\frac{m v^{2}}{2} = m g h $$


What two things do you need to measure?

```python
# Velocity and height. Note that since mass appears on both sides of the
# equation it cancels.
```

How might you find the uncertainty on each of these measurements?

```python
# I would perform an experiment to take these measurements many times, and from
# the distribution of the measurments I would report the uncertainty as 1 sigma
# (or 1 standard deviation) of the distribution
```

Below we have written two functions to help you measure the velocity of the ball and height. Running these functions would give you a 'measured' velocity and a 'measured 'height' to calculate g in your experiment.

```python
import numpy as np
```

```python
def measure_velocity():
    k=np.sqrt(2.*1.0*24.5)+np.random.normal(loc=0.0, scale=0.1, size=None)
    return k
```

```python
def measure_height():
    i=1.0+np.random.normal(loc=0.0, scale=0.001, size=None)
    return i
```

## Exercise:

Write a code which repeats this experiment, measuring the velocity and height 1000 times and plot a histogram of your results.

```python
import matplotlib.pyplot as plt

# Make empty lists
v = []
h = []
g = []

# Repeat the experiment 1000 times
for i in range(1000):
    v1 = measure_velocity()
    h1 = measure_height()
    g1 = (v1**2) / (2*h1)
    v.append(v1)
    h.append(h1)
    g.append(g1)

# Plotting
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist(v, bins=20)
plt.xlabel("Velocity (m/s)")
plt.ylabel("Frequency")

plt.subplot(1, 3, 2)
plt.hist(h, bins=20)
plt.xlabel("Height (m)")
plt.ylabel("Frequency")

plt.subplot(1, 3, 3)
plt.hist(g, bins=20)
plt.xlabel("g (m/s^2)")
plt.ylabel("Frequency")

plt.tight_layout()
```

What value of v and h would you report and therefore what value of g, and it's uncertainty, would you measure?  What planet are you performing this experiment on?

```python
# Mean gives the value of v and 1 std is the uncertainty
print(('v = %.3f +/- %.3f m/s') % (np.mean(v), np.std(v)))
```

```python
# Mean gives the value of h and 1 std is the uncertainty
print(('h = %.3f +/- %.3f m/s') % (np.mean(h), np.std(h)))
```

```python
# Mean gives the value of g and 1 std is the uncertainty
print(('g = %.2f +/- %.2f m/s') % (np.mean(g), np.std(g)))
print('I am on Jupiter since g is about 2.5 times larger than that of Earth')
```

This process of repeating a measurement is the real world equivalent of a monte carlo method. Monte Carlo is simply a method to explore the behaviour of a system by repeated random sampling of variables.

The Universe and our technology (e.g. how well our ruler measures) has provided the distribution and us repeating the measurement is the random sampling.
